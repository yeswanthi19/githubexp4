```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
```

```python
movies.head(2)
```

```python
genres = movies.groupby('Genre')
```

```python
# Applying builtin aggregation fuctions on groupby objects
genres.sum()
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
# find the top 3 genres by total earning
```

```python
# find the genre with highest avg IMDB rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find number of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
studentss=pd.read_csv('studentss.csv')
#studentss
```

```python
reg_month2=pd.read_csv('reg-month2.csv')
reg_month2.tail()
```

```python
studentss.merge(reg_month2,how='inner',on='student_id')
```

```python
courses=pd.read_csv('courses.csv')
```

```python
courses.tail()
```

```python
# left join
courses.merge(reg_month2,how='left',on='course_id')
```

```python
# right join
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([studentss,temp_df],ignore_index=True)
```

```python
students
```

```python
students.merge(reg_month2,how='right',on='student_id')
```

```python
reg_month2.tail()
```

```python
reg_month2.merge(students,how='left',on='student_id')
```

```python
# outer join
students.merge(reg_month2,how='outer',on='student_id').tail(10)
```

```python
# find total revenue generated
total = reg_month2.merge(courses,how='inner',on='course_id')['price'].sum()
total
```
